Code: 
**open("file name.extension","Mode")** 
- mode = read, write, overwrite...etc
  - r = reading mode
  - w = write mode
  - a = append
  - rb = read binary
  - wb = write binary
- if bu print a opened file, it will give the memory address and information about the file, not the data inside the file
- file.read() - this will print the data inside the file
- what if we want ot print only first line of the data?
  - file.readline()
  - this will print only first line
- what if i want o print 2nd line?
  - file.readline() - use this again 2nd time
- file.readline(4) : print 4 characters
- Writing data: **open("File name", "w")**
- if no file exists with that name then it will create a new file with that name
- file.write("Something") : this wil write a data in that file

- how to append data to new file?
  - open("file","a")
  - appends new data to a file

- this will work on image also!!
  - change the mode from a character mode --> Binary mode
  - rb = read binary
